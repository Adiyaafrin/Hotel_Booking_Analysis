### **Step 1: Handling missing values**

````python
# Write your code to make your dataset analysis ready.
Hotel_bookings_df.isnull().sum().sort_values(ascending=False)[:5]
````

**Earlier we had already mentioned that there was only Four columns that were having some Missing values in them and that was :(i).children,    (ii).agent,
                                               (iii).countries,   (iv).company.
                                               Now Let's replace those missing values in the above mentioned column with some meaningful value**


````python
# replacing null values in children column with 0 assuming that family had 0 children
# replacing null values in company and agent columns with 0 assuming those rooms were booked without company/agent

Hotel_bookings_df['children' ].fillna(0, inplace = True)
Hotel_bookings_df['company' ].fillna(0, inplace = True)
Hotel_bookings_df['agent' ].fillna(0, inplace = True)

# replacing null values in country column as 'Others'

Hotel_bookings_df['country'].fillna('Others', inplace = True)
````

### What all manipulations have you done and insights you found?

**In the 'company' and 'agent' columns, data represents company numbers and agent numbers.However, there are
instances where customers directly booked a hotel without involving any agent or hotel booking company.
To account for this, we will replace those instances with the value 0. Similarly, in the 'children' column,
there may be cases where the data was not entered. To handle this, we will replace any missing values in
the 'children' column with 0.**

````python
# Now there is no missing data in my DataFrame as all the missing values have been replaced with appropriate values, ensuring a more comprehensive
# and reliable dataset for further analysis and interpretation.
Hotel_bookings_df.isnull().sum().sort_values(ascending=False)[:5]
````

### **Step 3: Converting columns to proper dtype and format.**

````python
# Converting the data type of specific columns (children, agent, company) from 'float64' to 'int64':

Hotel_bookings_df[['children','company','agent']] = Hotel_bookings_df[['children','company','agent']].astype('int64')

# Let's convert the column name 'reservation_status_date' from 'Object' datatype to 'datetime':

Hotel_bookings_df['reservation_status_date'] = pd.to_datetime(Hotel_bookings_df['reservation_status_date'], format='%Y-%m-%d')
````

### **Step 4: Adding some columns.**

````python
# Adding a 'total_stay' column in the DataFrame, calculated as the sum of 'stays_in_week_nights' and 'stays_in_weekend_nights':
Hotel_bookings_df['Total_Stay']=Hotel_bookings_df['stays_in_week_nights']+Hotel_bookings_df['stays_in_weekend_nights']

# Adding a 'total_people' column in the DataFrame, calculated as the sum of 'adults', 'children', and 'babies'
Hotel_bookings_df['total_people'] = Hotel_bookings_df['adults'] + Hotel_bookings_df['children'] + Hotel_bookings_df['babies']
````

````python
# Chart - 7 visualization code
# Creating a datset of distribution channel name and count
dist_channel = Hotel_bookings_df['distribution_channel'].value_counts().reset_index().rename({'index':'distribution_channel','distribution_channel':'customer_sent'})
# Renaming the 'dist_channel' Index column to 'distribution_channel' and distribution_channel to 'customer_sent':
dist_channel=dist_channel.rename(columns={'index':'Channel name', 'distribution_channel':'Total_Customer_sent'})

#adding percentage columns to the distribution_channel_df
dist_channel['percentage']=round(dist_channel['Total_Customer_sent']*100/Hotel_bookings_df.shape[0],1)

# Plotting the chart values
plt.figure(figsize=(15,6))
my_explode = (0.02,0.02,0.02,0.02,0.02)
plt.pie(dist_channel['Total_Customer_sent'],explode=my_explode, labels=None, startangle = 50)

# Adding legends with percenatge using list comprehension
labels = [f'{l}, {s}%' for l, s in zip(dist_channel['Total_Customer_sent'].value_counts().index.tolist(), dist_channel['percentage'].values.tolist())]
plt.legend(bbox_to_anchor=(0.85, 1), loc='upper left', labels=labels)

plt.title('The booking distribution channels predominantly utilized by guests')
plt.axis('equal')
plt.show()
````

#**"Most Bookings Achieved: Top 5 Agents Highlighted"**

````python
# Chart - 10 visualization code
# Creating a dataset by grouping by agent column and it's count
# Just taken all the Top 5 Agent having more Booking_Count
agent_booking_count = Hotel_bookings_df['agent'].value_counts().head(10).reset_index().rename(columns={'index': 'Agent_Code', 'agent': 'Total_Booking_Count'}).sort_values(by='Total_Booking_Count', ascending=False)

# Extracting top 5 agents by booking count
top_agent = agent_booking_count[:5]

# Plotting the values
fig, ax = plt.subplots(figsize=(15, 7))
wedge_properties = {'linewidth': 1, 'edgecolor': "#39FF14"}
explode = (0.01, 0.01, 0.01, 0.01, 0.01)

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100. * np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Pie chart
wedges, texts, autotexts = ax.pie(top_agent['Total_Booking_Count'],
                                  autopct=lambda pct: func(pct, top_agent['Total_Booking_Count']),
                                  explode=explode,
                                  shadow=False,
                                  startangle=50,
                                  wedgeprops=wedge_properties)

# Adding legend
ax.legend(wedges, top_agent['Agent_Code'],
          title="agents",
          loc="upper left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
ax.set_title("The top five agents based on Booking", fontsize=20)

# Show chart
plt.axis('equal')
plt.show()
````

##### 1. Why did you pick the specific chart?

**Answer Here:**


A pie chart is chosen for the top 5 agents by booking due to its ability to visually represent the proportional contribution of each agent to the total bookings. The chart provides a clear and concise overview of the distribution, with percentages displayed, making it easy to identify and compare the booking volumes of the top agents at a glance.
